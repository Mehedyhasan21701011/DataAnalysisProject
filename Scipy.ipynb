```python
# Cluster
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.vq import kmeans2

a = np.random.multivariate_normal([0,6],[[2,1],[1,1.5]],size = 200)
# Creates 200 samples from a multivariate normal distribution with mean [0, 6] and covariance matrix [[2, 1], [1, 1.5]].
b = np.random.multivariate_normal([2,0],[[1,-1],[-1,3]],size = 400)
c = np.random.multivariate_normal([6,4],[[5,0],[0,1.2]],size = 250)
z = np.concatenate((a,b,c))
np.random.shuffle(z)
k = 3
centroids,labels = kmeans2(z,k)
# print('Centroids',centroids)
# print('Labels',labels)

cluster1 = z[labels==0]
cluster2 = z[labels==1]
cluster3 = z[labels==2]

# print('Cluster_1',cluster1)
# print('Cluster_2',cluster2)
# print('Cluster_3',cluster3)

plt.scatter(cluster1[:,0],cluster1[:,1],label = 'Cluster-1')
plt.scatter(cluster2[:,0],cluster2[:,1],label = 'Cluster-2')
plt.scatter(cluster3[:,0],cluster3[:,1],label = 'Cluster-3')

plt.scatter(centroids[:,0],centroids[:,1],marker = 'x',s = 100,label = 'Centroids',color = 'black')
plt.show()
```

```python
from scipy.interpolate import interp1d

x = np.arange(11)
y = np.array([2,1.9,1.7,1.5,0.5,0,2,0.9,0.8,0.7,0.6])

predict = interp1d(x,y,kind = 'quadratic')

x2 = np.linspace(0,12,25)
y2 = np.array([predict(x) for x in x2])

plt.plot(x,y,'o','.')
plt.plot(x2,y2,'ro')
plt.show()

```

```python
import numpy as np
from scipy.optimize import minimize

def objective(x):
  a = x[0]
  b = x[1]
  return a**2 + b**2

x0 = np.array([200,500])
res = minimize(objective,x0,method = 'SLSQP')
print(res)
```

```python
import numpy as np
from scipy.optimize import minimize

def objective(x):
  a = x[0]
  b = x[1]
  return a**2 + b**2


def constraint1(x):
  a = x[0]
  b = x[1]
  return a+b-100

def constraint2(x):
  a = x[0]
  b = x[1]
  return a*b - 100

cons = [{'type':'eq','fun':constraint1},{'type':'ineq','fun':constraint2}]

x0 = np.array([200,500])
res = minimize(objective,x0,method = 'SLSQP',constraints = cons)
print(res)
```

```python
import numpy as np
from scipy.optimize import minimize

def objective(x):
  a = x[0]
  b = x[1]
  return a**2 + b**2

x0 = np.array([200,500])
bound = np.array([10,50])
bounds = (bound,bound)
res = minimize(objective,x0,method = 'SLSQP',bounds = bounds)
print(res)
```
