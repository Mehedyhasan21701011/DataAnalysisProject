```python
# Rename columns name,make them small cases,replace np.spacing with underscore
file.rename(columns={'Order Id':'order_id'},inplace = True)
```

```python
file.columns = file.columns.str.lower()
```

```python
# @title region vs order_id

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(file['region'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(file, x='order_id', y='region', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```
